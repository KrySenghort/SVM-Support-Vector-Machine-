```python
import pandas as pd
```

```python
df = pd.read_csv('purchases.txt', '\t', names=['customer_id','purchase_amount','date_of_purchase', 'purchase_year'])
df['purchase_year'] = df.date_of_purchase.apply(lambda x: (x[:4]))

# pd.pivot_table(df.customer_id, df.purchase_amount) #, rownames=['customer_id'], colnames=['purchase_amount'])
pd.pivot_table(df, values='purchase_amount', index='customer_id').to_frame()
```

```python
purchases = pd.crosstab(df.customer_id, [df.purchase_year], rownames=['customer_id'], colnames=['purchase_year'])
purchases = purchases.applymap(lambda x: 1 if x>0 else 0)
purchases
```
